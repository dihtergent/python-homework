## ข้อ 1: สร้าง Dictionary และเข้าถึงค่า

สร้าง dictionary ชื่อ `student` ที่มีข้อมูลดังนี้:
- "name": "สมชาย"
- "age": 20
- "grade": "A"
- "major": "วิทยาศาสตร์คอมพิวเตอร์"

จากนั้นให้พิมพ์ชื่อและเกรดของนักเรียน

````python
# เขียนคำตอบของคุณที่นี่
student = {"name": "สมชาย",
           "age": 20,
           "grade": "A",
           "major": "วิทยาศาสตร์คอมพิวเตอร์"}

print("{} {}".format(student.get("name"),student.get("grade")))
````

## ข้อ 2: เพิ่ม แก้ไข และลบข้อมูลใน Dictionary

ให้ใช้ dictionary จากข้อ 1 แล้วทำดังนี้:
- เพิ่มคีย์ใหม่ "city" ที่มีค่า "กรุงเทพ"
- เปลี่ยนอายุเป็น 21
- ลบคีย์ "major" ออก

จากนั้นให้พิมพ์ dictionary ทั้งหมด

````python
# เขียนคำตอบของคุณที่นี่
student["city"] = "กรุงเทพ"
student["age"] = 21
student.pop("major")

print("{}".format(student))
````

## ข้อ 3: การวนลูปผ่าน Dictionary

สร้าง dictionary ชื่อ `prices` ที่เก็บราคาสินค้า:
```python
prices = {
```

## ข้อ 4: Tuple - สร้างและเข้าถึงค่า

สร้าง tuple ชื่อ `coordinates` ที่มีค่า (10, 20, 30)

จากนั้นให้พิมพ์:
- ค่าตัวแรก
- ค่าตัวสุดท้าย
- ค่าสามค่า (ใช้ slicing)
- จำนวนสมาชิกทั้งหมด

````python
# เขียนคำตอบของคุณที่นี่
coordinates = (10, 20, 30)
v = coordinates
print("{}\n{}\n{}\n{}\n".format(v[0],v[-1],v[0:3],v))
````

## ข้อ 5: Tuple ไม่สามารถเปลี่ยนแปลง (Immutable)

สร้าง tuple ชื่อ `colors` ที่มีค่า ("red", "green", "blue")

พยายามเปลี่ยนค่าตัวแรกเป็น "yellow" และบันทึกข้อความ error ที่ได้รับ

จากนั้นให้อธิบายว่าทำไม tuple ถึงไม่สามารถเปลี่ยนแปลงได้

````python
# เขียนคำตอบของคุณที่นี่
colors =  ("red", "green", "blue")
colors[0] = "yellow"

'''TypeError                                 Traceback (most recent call last)
Cell In[15], line 3
      1 # เขียนคำตอบของคุณที่นี่
      2 colors =  ("red", "green", "blue")
----> 3 colors[0] = "yellow"

TypeError: 'tuple' object does not support item assignment'''
````

## ข้อ 6: Dictionary ที่มีค่าเป็น Tuple

สร้าง dictionary ชื่อ `students_scores` ที่เก็บข้อมูลนักเรียน:
```python
students_scores = {
    "นักเรียน1": (85, 90, 88),
    "นักเรียน2": (75, 80, 78),
    "นักเรียน3": (95, 92, 94)
}
```

ให้คำนวณค่าเฉลี่ยคะแนนของนักเรียนแต่ละคน และพิมพ์ผลลัพธ์

````python
# เขียนคำตอบของคุณที่นี่
students_scores = {
    "นักเรียน1": (85, 90, 88),
    "นักเรียน2": (75, 80, 78),
    "นักเรียน3": (95, 92, 94)
}
s1 = students_scores["นักเรียน1"]
s2 = students_scores["นักเรียน2"]
s3 = students_scores["นักเรียน3"]
a = [
    (s1[0]+s1[1]+s1[2])/3,
    (s2[0]+s2[1]+s2[2])/3,
    (s3[0]+s3[1]+s3[2])/3
    ]
print("{}\n {}\n {}\n".format(a[0],a[1],a[2]))
   
````

## ข้อ 7: ค้นหาข้อมูลใน Dictionary

สร้าง dictionary ชื่อ `phone_book` ที่เก็บหมายเลขโทรศัพท์:
```python
phone_book = {
    "สมชาย": "081-234-5678",
    "สมหญิง": "089-876-5432",
    "สมศรี": "085-111-2222"
}
```

ให้สร้างฟังก์ชันที่ค้นหาหมายเลขโทรศัพท์ตามชื่อ และถ้าไม่พบให้แสดงข้อความ "ไม่พบข้อมูล"

## ข้อ 8: Tuple Unpacking

สร้าง tuple ชื่อ `person_info` ที่มีค่า ("วิชา", 22, "engineer")

ให้ใช้ tuple unpacking เพื่อกำหนดค่าให้กับตัวแปร name, age, และ job

จากนั้นพิมพ์ข้อมูลทั้งหมดในรูปแบบ "วิชา อายุ 22 ปี ประกอบอาชีพ engineer"

````python
# เขียนคำตอบของคุณที่นี่
person_info = ("วิชา", 22, "engineer")
(name,age,job) = person_info
print("{} อายุ{}ปี ประกอบอาชีพ {}".format(name,age,job))
````

## ข้อ 9: การรวม Dictionary

สร้าง dictionary 2 อัน:
- `person1 = {"name": "สมชาย", "age": 25}`
- `person2 = {"name": "สมหญิง", "age": 24}`

ให้รวม dictionary ทั้งสองเป็น dictionary เดียวชื่อ `all_people` และพิมพ์ผลลัพธ์

หมายเหตุ: ลองใช้วิธีต่างๆ เช่น update(), {**dict1, **dict2} เป็นต้น

````python
# เขียนคำตอบของคุณที่นี่
person1 = {"name": "สมชาย", "age": 25}
person2 = {"name": "สมหญิง", "age": 24}

all_people = {"p1":person1,"p2":person2}

print(all_people)
````

## ข้อ 10: โปรแกรมจัดการสินค้า (Challenge)

สร้างโปรแกรมจัดการสินค้าที่มีความสามารถดังนี้:

1. มี dictionary ชื่อ `inventory` ที่เก็บข้อมูลสินค้า: ชื่อสินค้า, จำนวน, และราคา (เก็บในรูป tuple)
2. สร้างฟังก์ชน `add_product()` เพื่อเพิ่มสินค้าใหม่
3. สร้างฟังก์ชั่น `update_quantity()` เพื่ออัปเดตจำนวนสินค้า
4. สร้างฟังก์ชั่น `get_total_value()` เพื่อคำนวณมูลค่ารวมของสินค้า
5. สร้างฟังก์ชั่น `show_inventory()` เพื่อแสดงรายการสินค้าทั้งหมด

ตัวอย่างการใช้งาน:
```
Inventory: {'น้ำตาล': (10, 50), 'แป้งมัน': (5, 100)}
Total value: 1000 บาท
```

````python
# เขียนคำตอบของคุณที่นี่

inventory = {'น้ำตาล': (10, 50), 'แป้งมัน': (5, 100)}

def add_product(name,qty,ppp):
    inventory[name] = (qty,ppp)
    return inventory
def update_quantity(name,qty):
    inventory[name] = (qty,inventory[name][1])
    return inventory

update_quantity("นํ้าตาล",3)
print(inventory)
````
